```python
import pandas as pd
import numpy as np

# Set random seed for consistency
np.random.seed(42)

# Number of rows
num_rows = 30

# Generate random data
data = {
    'Employee_ID': np.arange(1, num_rows + 1),  # Unique Employee ID
    'Name': np.random.choice(['Alice', 'Bob', 'Charlie', 'David', 'Emma', 'Frank', 'Grace', 'Hannah'], num_rows),
    'Age': np.random.randint(22, 60, num_rows),  # Age between 22 and 60
    'Salary': np.random.randint(35000, 120000, num_rows),  # Salary between 35k and 120k
    'Department': np.random.choice(['HR', 'IT', 'Finance', 'Marketing', 'Sales'], num_rows),
    'Joining_Date': np.random.choice(pd.date_range(start='2015-01-01', end='2023-12-31', freq='D'), num_rows)
}

# Create a DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
df.to_csv('data.csv')
```
